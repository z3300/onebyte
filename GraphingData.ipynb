```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

scores = pd.read_csv('/content/pagescraped.csv')
```

```python
scores['Fan Score'].replace(0, pd.np.nan, inplace=True)
scores = scores.rename(columns={'Resturant':'Restaurant'})

scores
```

```python
scores[scores.isnull().any(axis=1)]
```

```python
select = scores[['Date','Dave Score']]
select['Date'] = pd.to_datetime(select['Date'])
select = select.set_index('Date')
scores_monthly = select.loc['2016':].resample('M').mean().reset_index()
scores_monthly['Date_year'] = scores_monthly['Date'].dt.year
scores_monthly['Date'] = scores_monthly['Date'].dt.strftime('%b')
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
scores_monthly['Date'] = pd.Categorical(scores_monthly['Date'], categories=months, ordered=True)
scores_monthly = scores_monthly.sort_values(by='Date')
scores_monthly = scores_monthly.pivot(columns='Date', index='Date_year', values='Dave Score')
plt.figure(figsize=(10, 10))

sns.heatmap(scores_monthly, annot=True, fmt='.1f', cmap='OrRd')
plt.title('Average Dave Score by Month from 2016 onwards')
plt.xlabel('Month')
plt.ylabel('Year')
plt.savefig('heatmap', dpi=300)
plt.show()
```

```python
select = scores[['Date','Dave Score']]
select['Date'] = pd.to_datetime(select['Date'])
select = select.set_index('Date')
scores_monthly = select.loc['2016':].resample('M').count().reset_index()
scores_monthly['Date_year'] = scores_monthly['Date'].dt.year
scores_monthly['Date'] = scores_monthly['Date'].dt.strftime('%b')
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
scores_monthly['Date'] = pd.Categorical(scores_monthly['Date'], categories=months, ordered=True)
scores_monthly = scores_monthly.sort_values(by='Date')
scores_monthly = scores_monthly.pivot(columns='Date', index='Date_year', values='Dave Score')
plt.figure(figsize=(10, 8))

sns.heatmap(scores_monthly, annot=True, fmt='.1f', cmap='RdBu')
plt.title('Number of Reviews by Month from 2016 onwards')
plt.xlabel('Month')
plt.ylabel('Year')

plt.savefig('heatmap2', dpi=300)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Load your data into a pandas DataFrame

# Convert the 'Date' column to a datetime data type
scores['Date'] = pd.to_datetime(scores['Date'])

# Create a new column for the year of the review
scores['Year'] = scores['Date'].dt.year

# Use the groupby and count methods to count the number of reviews per year
review_counts = scores.groupby('Year').count()

sns.set(rc={'figure.figsize':(12,6)}) # Set the size of the figure

# Plot the number of reviews per year using sns.barplot
sns.barplot(x=review_counts.index, y=review_counts['Dave Score'])

# Add a title and axis labels
plt.title('Number of Reviews per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')

# Add the numbers on top of the bars
for i in range(len(review_counts)):
    plt.text(review_counts.index[i]-0.3, review_counts['Dave Score'][i]+5, str(review_counts['Dave Score'][i]), fontsize=8)

plt.savefig('countovertime', dpi=300)
# Show the plot
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Load your data into a pandas DataFrame

# Convert the 'Date' column to a datetime data type
scores['Date'] = pd.to_datetime(scores['Date'])

# Create a new column for the year of the review
scores['Year'] = scores['Date'].dt.year

# filter the dataframe for scores from 2016 and onwards
scores = scores.query('Year >= 2016')

# Group the data by year and calculate the average score for each year
average_scores = scores.groupby('Year')['Dave Score'].mean()

sns.set(rc={'figure.figsize':(12,6)}) # Set the size of the figure

# Plot the average score over each year using sns.lineplot
sns.lineplot(x=average_scores.index, y=average_scores, color="b")

# Add a title and axis labels
plt.title('Average Dave Score Over Each Year')
plt.xlabel('Year')
plt.ylabel('Average Score')

plt.savefig('averageovertime', dpi=300)
# Show the plot
plt.show()
```

```python
davescores = scores['Dave Score']

# Plot a kernel density plot of the scores
sns.kdeplot(davescores, shade=True)

# Add a title and axis labels
plt.title('Total Distribution of Dave Scores')
plt.xlabel('Score')
plt.ylabel('Density')

plt.savefig('Dstribution of Dave Scores', dpi=300)

# Show the plot
plt.show()
```

```python
dave_scores = scores['Dave Score']
fan_scores = scores['Fan Score']

plt.figure(figsize=(8, 6))
# Plot a kernel density plot of the Dave scores
sns.kdeplot(dave_scores, shade=True, color='b', label='Dave Scores')

# Plot a kernel density plot of the Fan scores
sns.kdeplot(fan_scores, shade=True, color='orange', label='Fan Scores')

# Add a title and axis labels
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Density')


plt.legend()

# Save the figure to a file
plt.savefig('score_distribution.png', dpi=300)

# Show the plot
plt.show()

```

```python
from scipy.stats import linregress
dave_scores = scores['Dave Score']
fan_scores = scores['Fan Score']


plt.scatter(dave_scores, fan_scores)


#calculate equation for trendline
z = np.polyfit(dave_scores, fan_scores, 1)
p = np.poly1d(z)

plt.plot(dave_scores, p(dave_scores))

# Add a title and axis labels
plt.title('Dave Scores vs Fan Scores')
plt.xlabel('Dave Scores')
plt.ylabel('Fan Scores')

# Save the figure to a file
plt.savefig('dave_vs_fan.png', dpi=300)
# Show the plot
plt.show()

```

```python
import plotly.express as px



# Calculate the average 'Dave Score' for each state
state_averages = scores.groupby('State')['Dave Score'].mean().reset_index()

state_averages = state_averages.dropna(subset=['State'])

# List of valid state codes
state_codes = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
               'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
               'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
               'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
               'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']

# Select only rows where the 'State' column is in the list of state codes
state_averages = state_averages.loc[state_averages['State'].isin(state_codes)]
fig = px.choropleth(state_averages,
                    locations='State',
                    color='Dave Score',
                    color_continuous_scale='agsunset',
                    hover_name='State',
                    locationmode='USA-states',
                    scope="usa") 
fig.update_layout(
    title_text = 'Average Dave Score by State',)
fig.show()
```
